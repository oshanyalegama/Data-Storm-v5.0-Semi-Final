```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_df.csv")##Loading the cleaned dataset
df.head()
```

## **Statistical analysis of dataset**

```python
df.describe()
```

```python
sns.pairplot(df)
plt.savefig('pairplot.eps', format='eps')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Creating a 3D Scatter Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot without color mapping (no clusters)
ax.scatter(df['average_monthly_spend'], df['average_monthly_visit_frequency'], df['average_monthly_basket_size'])

# Adding axis labels
ax.set_xlabel('average_monthly_spend')
ax.set_ylabel('average_monthly_visit_frequency')
ax.set_zlabel('average_monthly_basket_size')

# Adding title
ax.set_title('3D Scatter Plot of Features')

plt.savefig('3Dplot.eps', format='eps')

# Show plot
plt.show()
```

### From this it is clear that the data can be clustered into five segments. The cubical nature of the clusters also permits the use of K means clustering.
